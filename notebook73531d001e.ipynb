# ****Feature Engineering****

```python
data['Attrition'] = data['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
data.drop(['EmployeeNumber','Over18','StandardHours','EmployeeCount'],axis=1,inplace=True)
```

```python
x_cat = data.select_dtypes(include='object')
x_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
one_h_encode = OneHotEncoder()
x_cat = one_h_encode.fit_transform(x_cat).toarray()
```

```python
x_cat = pd.DataFrame(x_cat)
x_cat
```

```python
x_numerical = data.select_dtypes(exclude='object')
x_numerical.drop('Attrition',axis=1,inplace=True)
x_numerical
```

```python
new_data = pd.concat([x_cat,x_numerical],axis=1)
new_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
x = scalar.fit_transform(new_data)
```

```python
x
```

```python
y = data['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

# ****Modelling Phase****

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(f'Accuracy : {100 * accuracy_score(y_pred,y_test)}')
```

```python
print('Classification report')
print('======='*10)
print(classification_report(y_pred,y_test))
print('======='*10)
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(X_train,y_train)
```

```python
y_pred = randomforest.predict(X_test)
```

```python
print('Classification report')
print('======='*10)
print(classification_report(y_pred,y_test))
print('======='*10)
```
